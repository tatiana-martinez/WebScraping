```python
def getproduct():
    import time
    from pandas import json_normalize
    import requests as r
    import pandas as pd
    

    final_data = pd.DataFrame()
    for k in range(4): ## Since, I want to fetch data for 4 pages
        response = r.get(f'https://www.sarenza.com/store/product/gender-type/list/get?color=2&gender=1&material=559&material=591&material=790&material=80&sub_type=4&sub_type=6&sub_type=6&type=76&index={k}*100&count=100', headers=headers())
        result = response.json()
        flat_data = json_normalize(result)
        flat_data1 = json_normalize(flat_data['ProductVignetteList'][0])
        final_data = final_data.append(flat_data1)
        
        time.sleep(1) ## Here I'm asking my code to wait for 1 second after every request.
        
    return(final_data)
```

```python
def stock():
    from pandas import json_normalize
    
    dfproduct=getproduct()
    
    productstock=dfproduct.loc[dfproduct["Stock.InStockVariantList"].dropna().index].copy()
    
    list_of_jsons=productstock.apply((lambda x: [{j:k for j,k in list(i.items())+[("ProductColorId",x["ProductColorId"])]} 
                        for i in x["Stock.InStockVariantList"]]), axis=1).tolist()
    
    return(json_normalize([j for i in list_of_jsons for j in i]))  
```
